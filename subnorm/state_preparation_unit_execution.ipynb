# Unit execution for `subnorm/state_preparation.py`

```python
import sys
from pathlib import Path

import os

project_root = os.path.abspath('..')
if project_root not in sys.path:
    sys.path.insert(0, project_root)

import numpy as np

from subnorm.state_preparation import (
    random_subnormalized_state_sample,
    paper_style_state_preparation_data,
    recover_target_density_from_postselection,
)
```

```python
sample = random_subnormalized_state_sample(seed=7)

print('Normalized two-qubit state |psi_2> =')
print(sample.normalized_two_qubit_state)
print('Target subnormalised one-qubit state |psi_1> =')
print(sample.target_subnormalized_state)
print('Diagonal operator Sigma =')
print(sample.diagonal_operator)
print('||psi_1|| =', sample.state_norm)
print('Tr(|psi_1><psi_1|) =', sample.state_trace)
```

```python
print('Ancilla-|0> branch from ideal dilation =')
print(sample.dilation_result.ancilla_zero_branch)
print('Success probability =', sample.dilation_result.p_success)
print('Recovery error ||rho_recovered - rho_exact||_F =', sample.recovery_error_norm)
print('Recovered density equals exact density? ', np.allclose(
    sample.recovered_density_from_success_branch,
    sample.exact_density_matrix,
))
```

```python
rho_from_normalized_postselection = recover_target_density_from_postselection(
    sample.dilation_result.normalized_success_state,
    success_probability=sample.dilation_result.p_success,
    num_system_qubits=1,
)

print('Recovered from normalized postselection data =')
print(rho_from_normalized_postselection)
print('Matches exact density? ', np.allclose(rho_from_normalized_postselection, sample.exact_density_matrix))
```
